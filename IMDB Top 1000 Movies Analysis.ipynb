```python
df.drop(["Poster_Link","Overview","Star1","Star2","Star3","Star4"],axis=1,inplace=True)
```

```python
df.rename(columns={"Series_Title" : "Movie_Title"},inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Gross"] = df["Gross"].str.replace(",","")
df["Gross"] = df["Gross"].replace(np.nan,0)
df["Gross"] = df["Gross"].astype(int)
```

```python
df["Gross"] = df["Gross"].replace(0,df['Gross'].mean().astype(int))
```

```python
df["Meta_score"] = df["Meta_score"].replace(".","")
df["Meta_score"] = df["Meta_score"].replace(np.nan,0)
df["Meta_score"] = df["Meta_score"].astype(int)
```

```python
df["Meta_score"] = df["Meta_score"].replace(0,df['Meta_score'].mean().astype(int))
```

```python
#because the most movies is "U" :General Viewing for all ages
df["Certificate"].fillna('U',inplace=True)
```

```python
df["Runtime"] = df["Runtime"].str.replace(" min","")
df["Runtime"] = df["Runtime"].astype(int)
```
